## MDD definitions 

```python
sub_sample1_conditions = (
    (sub_metadata['2050-2.0']== 4) | (sub_metadata['2060-2.0']== 4 )
)
```

```python
sub_sample2_conditions = (
    (sub_metadata['2050-2.0']== 4 ) | (sub_metadata['2060-2.0']== 4 ) |
        (sub_metadata['2050-2.0']== 3 ) | (sub_metadata['2060-2.0']== 3)
)
```

```python
depression_ever_davis_conditions = (
    ((sub_metadata['20441-0.0']==1) | (sub_metadata['20446-0.0']==1))
    &
    ((sub_metadata['20436-0.0']==3) | (sub_metadata['20436-0.0']==4))
    &
    ((sub_metadata['20439-0.0']==2) | (sub_metadata['20439-0.0']==3))
    &
    ((sub_metadata['20440-0.0']==2) | (sub_metadata['20440-0.0']==3))
    &
    (((sub_metadata['20441-0.0']==1).astype('int') + (sub_metadata['20446-0.0']==1).astype('int') +
      (sub_metadata['20449-0.0']==1).astype('int') + (sub_metadata['20532-0.0']==1).astype('int') +
      (sub_metadata['20435-0.0']==1).astype('int') + (sub_metadata['20450-0.0']==1).astype('int') +
      (sub_metadata['20437-0.0']==1).astype('int') + (
       (sub_metadata['20536-0.0']==2) | (sub_metadata['20536-0.0']==1) | 
       (sub_metadata['20536-0.0']== 3)
      ).astype('int')) >= 5)
)
```

```python
sub_sample1 = sub_metadata.loc[sub_sample1_conditions]
sub_sample2 = sub_metadata.loc[sub_sample2_conditions]
depression_ever_davis = sub_metadata.loc[depression_ever_davis_conditions]

print(
    'sub_sample1:', len(sub_sample1),
    'sub_sample2:', len(sub_sample2),
    'depression_ever_davis:', len(depression_ever_davis) )
```

```python
guido_sub_sample_conditions1 = (depression_ever_davis_conditions & sub_sample1_conditions)
```

```python
guido_sub_sample_conditions2 = (depression_ever_davis_conditions & sub_sample2_conditions)
```

```python
guido_sub_sample = sub_metadata.loc[guido_sub_sample_conditions2]

print('guido_sub_sample:', len(guido_sub_sample))
```

```python
guido_sub_sample['MDD'] = 1
```
